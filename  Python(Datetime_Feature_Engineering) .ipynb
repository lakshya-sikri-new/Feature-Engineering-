# Convert to datetime

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['ship_date'] = pd.to_datetime(df['ship_date'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['order_date'].fillna(df['order_date'].mode()[0], inplace=True)
df['ship_date'].fillna(df['ship_date'].mode()[0], inplace=True)
```

# Extract Year, Month, Day, Weekday from order_date and ship_date


```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['order_day'] = df['order_date'].dt.day
df['order_weekday'] = df['order_date'].dt.weekday  # 0=Monday, 6=Sunday
df['order_quarter'] = df['order_date'].dt.quarter

df['ship_year'] = df['ship_date'].dt.year
df['ship_month'] = df['ship_date'].dt.month
df['ship_day'] = df['ship_date'].dt.day
df['ship_weekday'] = df['ship_date'].dt.weekday
df['ship_quarter'] = df['ship_date'].dt.quarter
df.head()
```

# Calculate the shipping duration in days

```python
df['shipping_duration'] = (df['ship_date'] - df['order_date']).dt.days
df.head()
```

```python
df['shipping_category'] = pd.cut(df['shipping_duration'], bins=[-1, 0, 2, 5, 10, float('inf')], 
                                 labels=['Same Day', '1-2 Days', '3-5 Days', '6-10 Days', '10+ Days'])
df.head()
```

# Weekend flag for orders and shipments


```python
df['order_is_weekend'] = df['order_date'].dt.weekday >= 5  # 5=Saturday, 6=Sunday

df['ship_is_weekend'] = df['ship_date'].dt.weekday >= 5
df.head()
```

```python
df = df.sort_values(by=['customer_name', 'order_date'])  # Sort by customer and date
df['days_since_last_order'] = df.groupby('customer_name')['order_date'].diff().dt.days
df.head()
```

# Define seasons based on month

```python
# Define seasons based on month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['order_season'] = df['order_month'].apply(get_season)
df.head()
```

# Holiday dates

```python
holidays = ['2024-01-01', '2024-12-25'] 
holidays = pd.to_datetime(holidays)

df['order_is_holiday'] = df['order_date'].isin(holidays)
df['ship_is_holiday'] = df['ship_date'].isin(holidays)
df.head(50)
```

```python
df['days_to_month_end'] = df['order_date'].apply(lambda x: (pd.Period(x, freq='M').end_time - x).days)
df.head()
```

```python
df['lag_1_day_sales'] = df['sales'].shift(1)
df['lag_7_day_sales'] = df['sales'].shift(7)
df.head()
```

# Week of the Month:

```python
df['order_week_of_month'] = df['order_date'].apply(lambda d: (d.day - 1) // 7 + 1)
df.head()
```

# Is Peak Season

```python
def is_peak_season(month):
    return month in [11, 12]

df['is_peak_season'] = df['order_month'].apply(is_peak_season)
df.head()
```
